```python
def op(data, op='mean'):
    if op == 'mean':
        return np.mean(data)
    elif op == 'std':
        return np.std(data)

t1, t2 = [], []
t3, t4 = [], []
for i in range(1, 88):
    cal = 'std'
    t1.append(op(train_X[train_y==0,i], cal))
    t2.append(op(train_X[train_y==1,i], cal))
    t3.append(op(test_X[test_y==0,i], cal))
    t4.append(op(test_X[test_y==1,i], cal))
```
