```python
import numpy as np
```

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('tyut.png')
```

```python
def imshow(img, color='rgb'):
    if color == 'bgr':
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    elif color == 'rgb':
        plt.imshow(img)
    elif color == 'gray':
        plt.imshow(img, 'gray')
    plt.show()
```

```python
imshow(img, 'bgr')
```

```python
def get_dark_channel(img, padding_width=5):
    dark = np.zeros((img.shape[0], img.shape[1]))
    r = np.pad(img[:, :, 0], ((padding_width, padding_width), (padding_width, padding_width)), mode='edge')
    g = np.pad(img[:, :, 0], ((padding_width, padding_width), (padding_width, padding_width)), mode='edge')
    b = np.pad(img[:, :, 0], ((padding_width, padding_width), (padding_width, padding_width)), mode='edge')
    img_pad = np.stack((r, g, b), axis=2)
    for x in range(dark.shape[0]):
        for y in range(dark.shape[1]):
            dark[x, y]=np.min(img_pad[x:x+2*padding_width, y:y+2*padding_width, :])
    return dark
```

```python
dark = get_dark_channel(img)
```

```python
imshow(dark, 'gray')
```
